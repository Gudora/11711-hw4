```python
import csv
headers = ["id", "topic", "summary"]

# Create a CSV file and write the headers
with open('/content/drive/MyDrive/summary.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(headers)

current_id = 0
current_diag = ""

# Assume you have some process where in each iteration you get a new row of data
for i in range(data.shape[0]):  # Example loop
    # New row of data for each iteration
    if data["Dialogue_ID"][i] != current_id:
      print(current_diag)
      print("---------------------")
      try:
        analysis = summary_topic_generator.generate_summary_and_topic(current_diag.strip())
        new_row = [current_id, analysis["topic"], analysis["summary"]]

        # Open the file in append mode and write the new row
        with open('/content/drive/MyDrive/summary.csv', 'a', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(new_row)
      except:
        pass
      current_id = data["Dialogue_ID"][i]
      current_diag = ""
    current_diag += f'#{data["Speaker"][i]}#: {data["Utterance"][i]}\n'
```
